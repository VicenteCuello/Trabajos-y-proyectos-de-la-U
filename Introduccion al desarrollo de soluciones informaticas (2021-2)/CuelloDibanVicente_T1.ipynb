```python
#Ejercicio 3
s=int(input("Salario:"))
while s<=0:
    s=s+1
    print("Error, ingrese un valor valido.")
    s=int(input("Salario:"))
if 0<s<500000:
    i=s*0.1
    res10=i+s
    r10=round(res10)
    print("Tu salirio a sido aumentado a $",r10,"pesos.")
elif 500000<=s<1000000:
    i=s*0.05
    res5=i+s
    r5=round(res5)
    print("Tu salirio a sido aumentado a $",r5,"pesos.")
else:
    print("No has recibido un aumento de salario.")
```
